```python
import pandas as pd
import numpy as np
import kagglehub
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from kagglehub import KaggleDatasetAdapter
```

```python
path = kagglehub.dataset_download("itszubi/fifa-player-performance-and-market-value")

print("Path to dataset files:", path)
```

```python
file_path = "fifa_player_performance_market_value.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "itszubi/fifa-player-performance-and-market-value",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
data = {
    'overall_rating': df['overall_rating'].head(20).tolist(),
    'market_value_million_eur': df['market_value_million_eur'].head(20).tolist(),
    'transfer_risk_level': df['transfer_risk_level'].head(20).tolist()
}

df_baru = pd.DataFrame(data)
print("DATA BARU DARI CSV FIFA PLAYER:")
print(df_baru)
print("\n")
```

```python
X = df_baru[['overall_rating', 'market_value_million_eur']]
y = df_baru['transfer_risk_level']
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Mapping transfer_risk_level ke angka:")
for i, level in enumerate(le.classes_):
    print(f"  {level} -> {i}")
print("\n")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("=" * 50)
print("HASIL PREDIKSI KNN")
print("=" * 50)
print("Data Testing (X_test):")
print(X_test)
print("\nTarget Asli (encoded):", y_test)
print("Target Prediksi:", y_pred)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nKeterangan: 0=Low, 1=Medium, 2=High")
```
